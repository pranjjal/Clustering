```python
import numpy as np
import pandas as pd
```

We took the data of final year students of a college having their iq and cgpa scores.

```python
df = pd.read_csv('student_clustering.csv')
print(df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

Trying elbow method to get how many clusters to form ?

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

We can see values of wcss is decreasing with increase in cluster numbers..

```python
plt.plot(range(1,11),wcss)
```

So we can see elbow point as 4.

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
X[y_means == 3,1]
```

```python
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],color='blue')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],color='red')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],color='green')
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],color='yellow')
```

K-Means on 3-D data

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
X
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2])
fig.show()
```

```python
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,21),wcss)
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

```python
df = pd.DataFrame()

df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred
```

```python
fig = px.scatter_3d(df,x='col1', y='col2', z='col3',color='label')
fig.show()
```

## AgglomerativeClustering

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
customer_data.shape
```

```python

customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
labels_=cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
